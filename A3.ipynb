# **Gender Identification on Twitter for Companies Advertisement According to Gender Specific**

## Problem Statement: 

Predict whether twitter author profile is male or female.

## Input:  

Twitter Text

## Output:  

Gender of human (male/female) 

## Goal: 

Learn from input to predict output

# Task_1

# Features Extraction Methods

## 1_Stylometry Based Features Extraction  

**Method**: I extracted 20 stylometry (‘,’,’.’,’(’,’)’,’!’,’-’,’?’,’%’,’&’,’#’,’_’,’=’,’A-Z’,’;’,’:’,’ space’,’a-z’,’/’,’0-9) based features count from each profile of male and female and save them in CSV file(each profile in single row with 20 features in twenty columns) using python script. 

## 2_Content Based Features Extraction

I wrote a python script for data cleaning (like special character, numbers, html decoding, @mention, URL links, hash tag etc.) of male and female profiles and after that I make directory folder for male clean profiles and same for females).  

## Word n-gram and Character n-gram

I extracted word n-gram and character n-gram from clean profiles data of male and females.

## Extraction Method

sklearn feature extraction TfidfVectorizer         

## Parameters of TfidfVectorizer

sublinear_tf=true, strip_accents=Unicode, stopwords=English, token pattern, n-gram_range, analyzer


![](https://drive.google.com/uc?export=view&id=12FxvX7J0An7BcRjNQVvD_8PEtbRFsr5O) 

## Features Selection Methods

### 1_Stylometry Based Features Selection 

I dropped some features (space, at the rate from 20 features) and increase efficiency score (10% to 15%) by selection of best stylometry based features.

### 2_Content Based Features Selection 

**Selection Method:** I used following methods for feature selection:
1-	Sklearn feature extraction TfidfVectorizer         
2-	Sklearn SelectKBest feature selection 
3-	Sklearn Chi2 feature selection 
I applied above mentioned modules for feature selection on extracted features of Word n-gram and Character n-gram. I selected highest scoring 1000 features for Word n-gram and Character n-gram.

```python
import glob
import csv
import pandas as pd
path_of_male_data="twitter-gender-corpus/male"
read_files_of_male = glob.glob(path_of_male_data+'/*.txt')
for i in read_files_of_male:
    with open(i, "rb") as infile_data_male:
        read_male_file=infile_data_male.read()
        yes=open("res.csv","w", encoding="utf8")
        bb=read_male_file.decode("utf-8")
        yes.write(bb)
        yes.close()
        yess=open("res.csv","r", encoding="utf8")
        bbb=yess.read()
        neww=open("comma.csv", "a", encoding="utf8")
        neww.write(str(bbb.count(',')))
        neww.write("\n")
        neww.close()
        neww=open("fullStop.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('.')))
        neww.write("\n")
        neww.close()
        neww=open("atThe.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('@')))
        neww.write("\n")
        neww.close()
        neww=open("openBr.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('(')))
        neww.write("\n")
        neww.close()
        neww=open("closeBr.csv", "a", encoding="utf8")
        neww.write(str(bbb.count(')')))
        neww.write("\n")
        neww.close()
        neww=open("exclamation.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('!')))
        neww.write("\n")
        neww.close()
        neww=open("Dash.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('-')))
        neww.write("\n")
        neww.close()
        neww=open("question.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('?')))
        neww.write("\n")
        neww.close()
        neww=open("percent.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('%')))
        neww.write("\n")
        neww.close()
        neww=open("andd.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('&')))
        neww.write("\n")
        neww.close()
        neww=open("hashtag.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('#')))
        neww.write("\n")
        neww.close()
        neww=open("underscore.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('_')))
        neww.write("\n")
        neww.close()
        neww=open("equal.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('=')))
        neww.write("\n")
        neww.close()
        neww=open("capitalletter.csv", "a", encoding="utf8")
        neww.write(str(sum(1 for c in bbb if c.isupper())))
        neww.write("\n")
        neww.close()
        neww=open("semiColon.csv", "a", encoding="utf8")
        neww.write(str(bbb.count(';')))
        neww.write("\n")
        neww.close()
        neww=open("Colon.csv", "a", encoding="utf8")
        neww.write(str(bbb.count(':')))
        neww.write("\n")
        neww.close()
        neww=open("sapceC.csv", "a", encoding="utf8")
        neww.write(str(bbb.count(' ')))
        neww.write("\n")
        neww.close()
        neww=open("smallletter.csv", "a", encoding="utf8")
        neww.write(str(sum(1 for c in bbb if c.islower())))
        neww.write("\n")
        neww.close()
        neww=open("forwardslash.csv", "a", encoding="utf8")
        neww.write(str(bbb.count('/')))
        neww.write("\n")
        neww.close()
        numeric = 0
        for i in bbb:  
            if i.isdigit():      
                numeric +=1         
                    
        neww=open("numberofD.csv", "a", encoding="utf8")
        neww.write(str(numeric))
        neww.write("\n")
        neww.close()
        neww=open("Gender.csv", "a", encoding="utf8")
        neww.write("1")
        neww.write("\n")
        neww.close()
        yesss=open("res.csv","w", encoding="utf8")
        yesss.truncate()
        yesss.close()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd

alldata=Male_Female_All.Authur_profile_Text

word_vectorizer = TfidfVectorizer(stop_words='english')

X= word_vectorizer.fit_transform(alldata)
```

```python
true_k = 2
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = word_vectorizer.get_feature_names()
```

```python
for i in range(true_k):
    print("Cluster %d: " % i)
    for ind in order_centroids[i, :1000]:
        print("%s" % terms[ind])
```
